```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
mpl.rcParams["font.family"] = "sans-serif"
import vapeplot
```

```python
xs = np.linspace(0, 25, 1000)

plt.plot(xs, np.zeros_like(xs), c="w")

plt.ylim(-5, 5)
plt.xlabel("sugar in grams")
plt.ylabel("utility");
```

# Linear preference for sugar

```python
plt.plot(xs, xs / 2.5 - 5, c="C1")

plt.ylim(-5, 5)
plt.xlabel("sugar in grams")
plt.ylabel("utility");
```

```python
plt.plot(xs, (10 - xs) / 2.5 + 1, c="C1")

plt.ylim(-5, 5)
plt.xlabel("sugar in grams")
plt.ylabel("utility");
```

# Optimum in the middle

```python
plt.plot(xs, 5 - np.abs(15 - xs) / 3 * 2, c="C1")

plt.ylim(-5, 5)
plt.xlabel("sugar in grams")
plt.ylabel("utility");
```

```python
plt.plot(xs, (np.sin(xs / 2.5) * 6 + xs) / 2.5 - 5.5, c="C1")

plt.ylim(-5, 5)
plt.xlabel("sugar in grams")
plt.ylabel("utility");
```
